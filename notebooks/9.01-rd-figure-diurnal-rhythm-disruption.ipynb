# Figure - Diurnal Rhythm Disruption Types
This code generates a figure illustrating different types of disruptions in diurnal rhythm, such as normal, prolonged, scaled, shifted, and unregular disruptions. Each type is represented by a subplot showing different patterns of time series data. The figure is saved as a PNG file in the specified directory.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from src.config import FIGURES_DIR

idx = np.linspace(0, 10, 100)

df = pd.DataFrame({'line1': np.sin(idx), 'line2': np.sin(idx/2), 'line3': np.sin(idx)*0.5, 'line5': np.sin(idx) * np.cos((idx/0.5)-2)}, index=idx)
df['line4'] = df['line1'].shift(20)
noise = np.random.uniform(-0.05, 0.05, len(idx))
fig, ax = plt.subplots(figsize=(10,2), ncols=5)
ax[0].plot(pd.concat([df['line1'], df['line1'] + noise +0.05], axis=1))
ax[0].set_title('(a) Normal')
ax[1].plot(df[['line1', 'line2']])
ax[1].set_title('(b) Prolonged')
ax[2].plot(df[['line1', 'line3']])
ax[2].set_title('(c) Scaled')
ax[3].plot(df[['line1', 'line4']])
ax[3].set_title('(d) Shifted')
ax[4].plot(df[['line1', 'line5']])
ax[4].set_title('(e) Irregular')
for a in ax:
    a.set_xticks([])
    a.set_yticks([])
    a.grid(False)
plt.tight_layout()
plt.savefig(FIGURES_DIR / 'disruption_types.png', dpi=300, bbox_inches='tight')
plt.show()
```
